```python
year = int(input())

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("윤년입니다.")
else:
    print("윤년이 아닙니다.")
```

```python
num1 = int(input())
num2 = int(input())
oper = input("연산자를 적어주세요.")

if oper == "+":
    print(num1 + num2)
elif oper == "-":
    print(num1 - num2)
elif oper == "*":
    print(num1 * num2)
elif oper == "/":
    print(num1 / num2)
else:
    print("연산자가 아닙니다.")
```
