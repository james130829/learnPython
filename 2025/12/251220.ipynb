```python
siu = []
number = [5,-3,0,8,-2]
for n in number:
    siu.append(n * -1)
print(siu)
```

```python
number = [5,-3,0,8,-2]
siu = [n * -1 for n in number]
print(siu)
```

```python
number = [3,6,9,12,15,4,8,11,5]
even = 0
odd = 0
for n in number:
    if n % 2 == 0:
        even = even + 1
    if n % 2 != 0:
        odd = odd + 1
if even > odd:
    print("짝수가 더 많음")
if odd > even:
    print("홀수가 더 많음")
```
