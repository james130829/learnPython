```python
def num (A):
    if A == 0:
        print("zero")
    elif A % 2 != 0:
        print("odd")
    elif A % 2 == 0:
        print("even")
A = int(input())
num(A)
```

```python
def even (A):
    num = []
    for n in A:
        if n % 2 == 0:
            num.append(n)
    return (num)

S = [1,2,3,4,5,6]
result=even(S)
print(even(S))
print(result)
```

```python
def is_pass (scores):
    if scores >= 60:
        return True
    else:
        return False
    
def result(score):
    jum = 0
    for s in score:
        if is_pass(s) == True:
            jum = jum + 1
    return jum

score = [80,55,90,40,70]
result(score)
```

```python
def check_answer ():
    answer = []
    numbers = [3, 7, 2]
    for r in range(len(numbers)):
        player = int(input())
        answer.append(player)
    if numbers == answer:
        print("SUCCESS")
    else:
        print("FAIL")

check_answer()
```

```python
def check_stage ():
    answer = []
    numbers = [1, 2, 3]
    for r in range(len(numbers)):
        player = int(input())
        answer.append(player)
    if numbers == answer:
        print(numbers, answer)
    else:
        print("FAIL")

check_stage()
```
