```python
def type_Triangle (a,b,c):
    e = 0
    if a < b:
        e = b
        b = a
        a = e
    if a < c:
        e = c
        c = a
        a = e
    if a<(b+c):
        if a == b == c:
            print(1)
        elif a == b or c == b or a == c:
            print(2)
        else:
            print(3)
    else:
        print(0)



a = int(input())
b = int(input())
c = int(input())
type_Triangle(a,b,c)
```
