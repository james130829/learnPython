```python
def countSizes ():
    LMN = ["XS","XS","XS","XS","M","S","XXL"]
    unifom = [0,0,0,0,0,0]
    for l in LMN:
        if l == "XS":
            unifom[0]+=1
        if l == "S":
            unifom[1]+=1
        if l == "M":
            unifom[2]+=1
        if l == "L":
            unifom[3]+=1
        if l == "XL":
            unifom[4]+=1
        if l == "XXL":
            unifom[5]+=1
    return unifom
countSizes ()
```

```python
def moreNumber ():
    num = [2,3,5,7,10,15]
    even = 0
    odd = 0
    for n in num:
        if n % 2 == 0:
            even += 1
        elif n % 2 == 1:
            odd += 1
    if even < odd:
        print("odds")
    elif odd < even:
        print("evens")
    elif odd == even:
        print("same")
moreNumber()
```
