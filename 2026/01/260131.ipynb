```python
def checkPassword (a):
    b = 0
    c = 0
    d = 0
    e = True
    for i in a:
        if i.isupper():
            b += 1
        elif i.islower():
            c += 1
        elif i.isdigit():
            d += 1
        elif i not in "@$&*!":
            # 해당 문자 in 문자열
            e += False
    if b >= 1 and c >= 2 and d >= 2 and e :
        print("비밀번호 사용가능")
    else:
        print("사용불가")
x = input()
checkPassword(x)
```
